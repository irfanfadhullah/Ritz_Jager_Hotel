Jika berjenis bukan normal distribution, maka diisi dengan median, note jika missing value tidak lebih dari 40-50%

```python
#Precentage the missing values
data.isnull().sum() * 100 / len(data)
```

AS you can see from the result above, the precentage of missing values of company  column is about 94%, so we must drop the column or we can ignored that column, because it is not fulfill the criteria. However, the agent, children and country origin column we can fill the missing values with the mean, median or mode.

```python
fillagent=data['agent'].median()
fillchild=data['children'].median()
data['agent']=data['agent'].fillna(fillagent)
data['children']=data['children'].fillna(fillchild)
data
```

```python
fillcountry=data['country_origin'].mode()[0] #You must add [0] that indicate you fill with the first value in index of mode
data['country_origin']=data['country_origin'].fillna(fillcountry)
```

```python
#1. Where do the guests come from? (assumption the guests with no cancel booking)
```

```python
datacrealguest=datac[datac['is_canceled']==0]
country=datacrealguest.country_origin.value_counts()
country=pd.DataFrame(country)
country

country= pd.DataFrame(country).reset_index()
country.rename(columns={"index": "countrycode", "country_origin":"guest"}, inplace=True)

#This code i got from the internet to convert from ISO country code to Country Name
import pycountry

#list the name country
list_alpha_2 = [i.alpha_2 for i in list(pycountry.countries)]
list_alpha_3 = [i.alpha_3 for i in list(pycountry.countries)]    

def country_flag(df):
    if (len(df['countrycode'])==2 and df['countrycode'] in list_alpha_2):
        return pycountry.countries.get(alpha_2=df['countrycode']).name
    elif (len(df['countrycode'])==3 and df['countrycode'] in list_alpha_3):
        return pycountry.countries.get(alpha_3=df['countrycode']).name
    else:
        return 'Invalid Code'

country['countrycode']=country.apply(country_flag, axis = 1)
country
```

<h2>As we can see from the result above, the guest is from 165 different country, with Portugal as the most guest come from.</h2>

```python
#2. How much do guests pay for a room per night? 
```

```python
guest=datacrealguest[(datacrealguest.loc[:,'adults']>0)&(datacrealguest.loc[:,'children']>0)]
guest['hotel_type']
```

```python
#create and store the total of guest from adult and children category to total column
guest.loc[:,'total']=guest.loc[:,'adults']+guest.loc[:,'children']

#create and store the total guest pay from adr column and divided by total of adult and children category
guest.loc[:,'paypernight']=guest.loc[:,'adr']/guest.loc[:,'total']
guest
```

```python
#divided hotel by type, as we can see from the dataset, there are 2 type of hotel, city hotel and resort hotel
cityhotel=guest[guest['hotel_type']=='City Hotel']
resorthotel=guest[guest['hotel_type']=='Resort Hotel']

#calculate the mean of cost of City Hotel
citycost=cityhotel['paypernight'].mean()
citycost=round(citycost)

#calculate the mean cost of Resort Hotel
resortcost=resorthotel['paypernight'].mean()
resortcost=round(resortcost)
print("The average guests pay for a room at City Hotel at $" +str(citycost)+
      '\nAnd the average guests pay for a room at Resort Hotel at $' +str(resortcost))
```
